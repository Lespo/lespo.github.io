```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configs o the plots we will create

# Read In Data

df = pd.read_csv(r'C:\Users\cyrix\Documents\Portfolio\Project 3\movies.csv')
```

```python
# Display Data

df.head()
```

```python
# Take a look at missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types for our Columns

df.dtypes
```

```python
# Change Data Types For Columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df.dtypes
```

```python
#Correct Year

df['year_correct'] = df['year'].astype(str).str[:4]
df
```

```python
#Sort by gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Set To See all rows

pd.set_option('display.max_rows', 10)
```

```python
# Drop any Duplicates

#df.drop_duplicates()
```

```python
# Scatter Plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
# Plot Budget vs Gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
plt.show()
```

```python
# Corrolation

df.corr(method='pearson', numeric_only=True) 
```

```python
# Correlation Matrix

correlation_matrix = df.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')



plt.show()
```
